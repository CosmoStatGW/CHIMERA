```python
from functools import partial
import jax
jax.config.update('jax_platform_name', 'cpu')
import jax.numpy as jnp

@jax.jit
def E_at_z(z, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0.):   

    Ode0 = 1. - Om0 - Ok0 - Or0
    return jnp.sqrt(
        Om0 * (1. + z)**3 + 
        Ok0 * (1. + z)**2 + 
        Or0 * (1. + z)**4 + 
        Ode0 * (1. + z)**(3. * (1. + w0 + wa)) * jnp.exp(-3. * wa * z / (1. + z))
    )

@partial(jax.jit, static_argnames = ['res'])
def int_invE_at_z(z, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0., res=500):
    
    max_z    = jnp.max(jnp.atleast_1d(z))
    z_values = jnp.linspace(0.0, max_z,res)
    dz = z_values[1] - z_values[0]

    e_vals = 1. / E_at_z(z_values, Om0, Ok0, Or0, w0, wa)
    
    cum_integral_values = jnp.empty_like(z_values)
    cum_integral_values = cum_integral_values.at[0].set(0)
    cum_integral_values = cum_integral_values.at[1:].set(jnp.cumsum(e_vals[:-1] + e_vals[1:])/2 *dz)

    #cum_integral_values = jnp.cumsum((e_vals[:-1] + e_vals[1:]) * dz / 2)

    return jnp.interp(z, z_values, cum_integral_values)
         
@jax.jit
def dCr_at_z(z, H0=70, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0.):
    
    int_invEz = int_invE_at_z(z, Om0, Ok0, Or0, w0, wa)   
    dH  = 299792.458e-3 /H0
    dCr = dH * int_invEz
    return dCr 

@jax.jit
def dCt_at_z(z, H0=70, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0., distances = None):
        
    if distances:
        assert z.shape == distances.shape
        return distances/(1.+z)
    
    dCr = dCr_at_z(z,H0,Om0,Ok0,Or0,w0,wa)
    sqrtOk0 = jnp.sqrt(jnp.abs(Ok0))
    dH      = 299792.458 / H0
    
    dCt = jnp.where(Ok0 == 0.0,
                    dCr,
                    jnp.where(Ok0 > 0.0,
                              (dH / sqrtOk0) * jnp.sinh(sqrtOk0 * dCr / dH),
                              (dH / sqrtOk0) * jnp.sin(sqrtOk0 * dCr / dH)
                             )
                   )
    return dCt

@jax.jit
def dA_at_z(z, H0=70, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0., distances = None):
    
    if distances:
        assert z.shape == distances.shape
        return distances/(1.+z)**2
    
    dCt = dCt_at_z(z, H0, Om0, Ok0, Or0, w0, wa) 
    dA  = dCt/(1.+z)
    return dA

@jax.jit
def dL_at_z(z, H0=70, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0.):    
    dCt = dCt_at_z(z, H0, Om0, Ok0, Or0, w0, wa)
    dL  = dCt*(1.+z)
    return dL

@jax.jit
def ddL_at_z(z, H0=70, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0., distances = None):
    dH = 299792.458e-3 / H0
    
    if distances is None:
        E_z    = E_at_z(z, Om0, Ok0, Or0, w0, wa)       
        dCt    = dCt_at_z(z, H0, Om0, Ok0, Or0, w0, wa)
        ddL    = dCt + (dH/E_z)*(1.+z)
    else:
        assert z.shape == distances.shape
        E_z = E_at_z(z, Om0, Ok0, Or0, w0, wa) 
        dCt = distances/(1+z)
        ddL = dCt + (dH/E_z)*(1.+z)
        
    return ddL

@jax.jit
def Vc_at_z(z, H0=70, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0., distances = None):

    dH = 299792.458e-3 / H0 
    if distances is None:
        dCt = dCt_at_z(z, H0, Om0, Ok0, Or0, w0, wa)
    else:
        assert z.shape == distances.shape
        dCt = distances/(1.+z)

    Vc = jnp.where(Ok0 == 0.0,
                   4.*jnp.pi*dCt**3 / 3.,
                   jnp.where(Ok0 > 0.0,
                             (4. * jnp.pi * dH**3 / (2. * Ok0)) * (
                                 (dCt / dH) * jnp.sqrt(1 + Ok0 * dCt**2 / dH**2)
                                 - jnp.arcsinh(jnp.sqrt(Ok0) * dCt / dH) / jnp.sqrt(Ok0)),
                             (4. * jnp.pi * dH**3 / (2. * Ok0)) * (
                                 (dCt / dH) * jnp.sqrt(1 + Ok0 * dCt**2 / dH**2)
                                 - jnp.arcsin(jnp.sqrt(abs(Ok0)) * dCt / dH) / jnp.sqrt(abs(Ok0)))
                             )
                   )
    return Vc

@jax.jit
def dVc_at_z(z, H0=70, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0., distances = None):
    
    dH = 299792.458e-3 / H0
    E_z = E_at_z(z, Om0, Ok0, Or0, w0, wa)
            
    if distances is None:
        dA = dA_at_z(z, H0, Om0, Ok0, Or0, w0, wa)
    else:
        assert z.shape == distances.shape
        dA = distances/(1.+z)**2

    dVc = dH * dA**2 / E_z

    return dVc

@partial(jax.jit, static_argnames = ['max_z', 'res'])
def dL2z(dLs,H0=70, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0., max_z = 10., res = 250):
    z_values = jnp.linspace(0.0, max_z, res) 
    dL_values = dL_at_z(z_values, H0, Om0, Ok0, Or0, w0, wa)
    return jnp.interp(dLs, dL_values, z_values)

#########################################################################################

# mg functions

@jax.jit
def Xi_at_z(z, Xi0, n):
    return Xi0 + (1.-Xi0)/(1.+z)**n

@jax.jit
def dGW_at_z(z, H0=70, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0., Xi0 = 1., n = 2.5):    
    dLz = dL_at_z(z, H0, Om0, Ok0, Or0, w0, wa) 
    Xiz = Xi_at_z(z, Xi0, n)
    return dLz*Xiz

@jax.jit 
def ddGW_at_z (z, H0=70, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0., Xi0 = 1., n = 2.5, distances = None):
    
    if distances:
        assert z.shape == distances.shape
        dLz = distances
    else:
        dLz  = dL_at_z(z, H0, Om0, Ok0, Or0, w0, wa)
    ddLz = ddL_at_z(z, H0, Om0, Ok0, Or0, w0, wa, distances = dLz)
    
    Xiz  = Xi_at_z(z, Xi0, n)
    dXiz = (Xi0-1.)/(1.+z)**(n+1)
            
    return ddLz*Xiz + dLz*dXiz

@partial(jax.jit, static_argnames = ['max_z', 'res'])
def dGW2z(dGWs,H0=70, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0., Xi0=1., n=2.5, max_z = 10., res = 250):
    z_values = jnp.linspace(0.0, max_z, res) 
    dGW_values = dGW_at_z(z_values, H0, Om0, Ok0, Or0, w0, wa, Xi0, n)
    return jnp.interp(dGWs, dGW_values, z_values)
```

```python
from functools import partial
import jax
jax.config.update('jax_platform_name', 'cpu')
jax.config.update("jax_enable_x64", True)
import jax.numpy as jnp

@jax.jit
def E_at_z(z, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0.):   

    Ode0 = 1. - Om0 - Ok0 - Or0
    return jnp.sqrt(
        Om0 * (1. + z)**3 + 
        Ok0 * (1. + z)**2 + 
        Or0 * (1. + z)**4 + 
        Ode0 * (1. + z)**(3. * (1. + w0 + wa)) * jnp.exp(-3. * wa * z / (1. + z))
    )

@partial(jax.jit, static_argnames = ['res'])
def int_invE_at_z(z, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0., res=500):
    
    max_z    = jnp.max(jnp.atleast_1d(z))
    z_values = jnp.linspace(0.0, max_z,res)
    dz = z_values[1] - z_values[0]

    e_vals = 1. / E_at_z(z_values, Om0, Ok0, Or0, w0, wa)
    
    cum_integral_values = jnp.empty_like(z_values)
    cum_integral_values = cum_integral_values.at[0].set(0)
    cum_integral_values = cum_integral_values.at[1:].set(jnp.cumsum(e_vals[:-1] + e_vals[1:])/2 *dz)

    #cum_integral_values = jnp.cumsum((e_vals[:-1] + e_vals[1:]) * dz / 2)

    return jnp.interp(z, z_values, cum_integral_values)
         
@jax.jit
def dCr_at_z(z, H0=70, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0.):
    
    int_invEz = int_invE_at_z(z, Om0, Ok0, Or0, w0, wa)   
    dH  = 299792.458e-3 /H0
    dCr = dH * int_invEz
    return dCr 

@jax.jit
def dCt_at_z(z, H0=70, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0., distances = None):
        
    if distances:
        assert z.shape == distances.shape
        return distances/(1.+z)
    
    dCr = dCr_at_z(z,H0,Om0,Ok0,Or0,w0,wa)
    sqrtOk0 = jnp.sqrt(jnp.abs(Ok0))
    dH      = 299792.458 / H0
    
    dCt = jnp.where(Ok0 == 0.0,
                    dCr,
                    jnp.where(Ok0 > 0.0,
                              (dH / sqrtOk0) * jnp.sinh(sqrtOk0 * dCr / dH),
                              (dH / sqrtOk0) * jnp.sin(sqrtOk0 * dCr / dH)
                             )
                   )
    return dCt

@jax.jit
def dA_at_z(z, H0=70, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0., distances = None):
    
    if distances:
        assert z.shape == distances.shape
        return distances/(1.+z)**2
    
    dCt = dCt_at_z(z, H0, Om0, Ok0, Or0, w0, wa) 
    dA  = dCt/(1.+z)
    return dA

@jax.jit
def dL_at_z(z, H0=70, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0.):    
    dCt = dCt_at_z(z, H0, Om0, Ok0, Or0, w0, wa)
    dL  = dCt*(1.+z)
    return dL

@jax.jit
def ddL_at_z(z, H0=70, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0., distances = None):
    dH = 299792.458e-3 / H0
    
    if distances is None:
        E_z    = E_at_z(z, Om0, Ok0, Or0, w0, wa)       
        dCt    = dCt_at_z(z, H0, Om0, Ok0, Or0, w0, wa)
        ddL    = dCt + (dH/E_z)*(1.+z)
    else:
        assert z.shape == distances.shape
        E_z = E_at_z(z, Om0, Ok0, Or0, w0, wa) 
        dCt = distances/(1+z)
        ddL = dCt + (dH/E_z)*(1.+z)
        
    return ddL

@jax.jit
def Vc_at_z(z, H0=70, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0., distances = None):

    dH = 299792.458e-3 / H0 
    if distances is None:
        dCt = dCt_at_z(z, H0, Om0, Ok0, Or0, w0, wa)
    else:
        assert z.shape == distances.shape
        dCt = distances/(1.+z)

    Vc = jnp.where(Ok0 == 0.0,
                   4.*jnp.pi*dCt**3 / 3.,
                   jnp.where(Ok0 > 0.0,
                             (4. * jnp.pi * dH**3 / (2. * Ok0)) * (
                                 (dCt / dH) * jnp.sqrt(1 + Ok0 * dCt**2 / dH**2)
                                 - jnp.arcsinh(jnp.sqrt(Ok0) * dCt / dH) / jnp.sqrt(Ok0)),
                             (4. * jnp.pi * dH**3 / (2. * Ok0)) * (
                                 (dCt / dH) * jnp.sqrt(1 + Ok0 * dCt**2 / dH**2)
                                 - jnp.arcsin(jnp.sqrt(abs(Ok0)) * dCt / dH) / jnp.sqrt(abs(Ok0)))
                             )
                   )
    return Vc

@jax.jit
def dVc_at_z(z, H0=70, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0., distances = None):
    
    dH = 299792.458e-3 / H0
    E_z = E_at_z(z, Om0, Ok0, Or0, w0, wa)
            
    if distances is None:
        dA = dA_at_z(z, H0, Om0, Ok0, Or0, w0, wa)
    else:
        assert z.shape == distances.shape
        dA = distances/(1.+z)**2

    dVc = dH * dA**2 / E_z

    return dVc

@partial(jax.jit, static_argnames = ['max_z', 'res'])
def dL2z(dLs,H0=70, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0., max_z = 10., res = 250):
    z_values = jnp.linspace(0.0, max_z, res) 
    dL_values = dL_at_z(z_values, H0, Om0, Ok0, Or0, w0, wa)
    return jnp.interp(dLs, dL_values, z_values)

#########################################################################################

# mg functions

@jax.jit
def Xi_at_z(z, Xi0, n):
    return Xi0 + (1.-Xi0)/(1.+z)**n

@jax.jit
def dGW_at_z(z, H0=70, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0., Xi0 = 1., n = 2.5):    
    dLz = dL_at_z(z, H0, Om0, Ok0, Or0, w0, wa) 
    Xiz = Xi_at_z(z, Xi0, n)
    return dLz*Xiz

@jax.jit 
def ddGW_at_z (z, H0=70, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0., Xi0 = 1., n = 2.5, distances = None):
    
    if distances:
        assert z.shape == distances.shape
        dLz = distances
    else:
        dLz  = dL_at_z(z, H0, Om0, Ok0, Or0, w0, wa)
    ddLz = ddL_at_z(z, H0, Om0, Ok0, Or0, w0, wa, distances = dLz)
    
    Xiz  = Xi_at_z(z, Xi0, n)
    dXiz = (Xi0-1.)/(1.+z)**(n+1)
            
    return ddLz*Xiz + dLz*dXiz

@partial(jax.jit, static_argnames = ['max_z', 'res'])
def dGW2z(dGWs,H0=70, Om0=0.3, Ok0=0., Or0=0., w0=-1., wa=0., Xi0=1., n=2.5, max_z = 10., res = 250):
    z_values = jnp.linspace(0.0, max_z, res) 
    dGW_values = dGW_at_z(z_values, H0, Om0, Ok0, Or0, w0, wa, Xi0, n)
    return jnp.interp(dGWs, dGW_values, z_values)

############################################################################

# classes, wrap only methods used in GW cosmo

import logging
logger = logging.getLogger(__name__)

class flrw:
    def __init__(self, fiducial_lambda_c=None):
        self.name = 'flrw'
        self.lambda_c = { # default values
            'H0': 70.,
            'Om0': 0.3,
            'Ok0': 0.0,
            'Or0': 0.0,
            'w0': -1.0,
            'wa': 0.0
        }
        self.cosmo_keys = self.lambda_c.keys()
        self.vectorize_computation = False
        self.update_fiducial(fiducial_lambda_c)     
        logger.info(f'Created `{self.name}` model with fiducial cosmological parameters: {self.fiducial_lambda_c}')

    def update_fiducial(self, fiducial_lambda_c=None):
        if fiducial_lambda_c:
            # check if fiducial_lambda_c contains correct keys
            invalid_keys = [k for k in fiducial_lambda_c.keys() if k not in self.cosmo_keys]
            if invalid_keys:
                raise ValueError(f"`fiducial_lambda_c` contains invalid keys {invalid_keys}. Valid keys are {self.cosmo_keys}.")
            else:
                if all(not isinstance(v, jnp.ndarray) for v in fiducial_lambda_c.values()):
                    self.lambda_c.update(fiducial_lambda_c)
                else:
                    raise ValueError("Use only `float` parameters to initialize the class.")
            self.fiducial_lambda_c = self.lambda_c.copy()   # store fiducials 
        else:
            self.fiducial_lambda_c = self.lambda_c.copy()   # store fiducials 
        
    def update_params(self, lambda_c):
        
        # check if lambda_c contains correct keys
        invalid_keys = [k for k in lambda_c.keys() if k not in self.cosmo_keys]
        if invalid_keys:
            raise ValueError(f"`lambda_c` contains invalid keys {invalid_keys}. Valid keys are {self.cosmo_keys}.")   
            
        up_params            = {k:jnp.atleast_1d(lambda_c[k]) for k in lambda_c}
        keys_to_update       = up_params.keys()
        shape_updated_params = jnp.array([up_params[k].shape[0] for k in keys_to_update])
        
        if not jnp.all(jnp.isclose(shape_updated_params, shape_updated_params[0])):
            raise ValueError("All parameter arrays must have the same length")
        
        self.vectorize_computation = shape_updated_params[0] > 1
        
        if shape_updated_params[0] == 1:
            self.lambda_c.update({k:float(up_params[k][0]) for k in up_params}) 
        else:
            # broadcast not updated keys to have the same shape of the updated ones 
            # and values equal to the fiducial ones
            non_up_params = {k:jnp.full(shape_updated_params[0], self.fiducial_lambda_c[k]) for k in self.cosmo_keys if k not in keys_to_update}
            self.lambda_c.update(up_params)
            self.lambda_c.update(non_up_params)

    def z_from_dGW(self, dLs, max_z = 10., res = 250):
        if self.vectorize_computation:
            # dLs will aways have shape (), (nsamples,) or (nevents,nsamples)
            # need only vectorization on cosmo params arrays
            # output has shape (nparams,), (nparams,nsamples) or (nparams,nevents,nsamples)
            return jax.vmap(dL2z, in_axes=(None,0,0,0,0,0,0,None,None))(dLs, *self.lambda_c.values(), max_z, res)
        return dL2z(dLs, *self.lambda_c.values(), max_z , res)
        
    def dVdz_at_z(self, z, distances=None):
        if self.vectorize_computation:
            if len(jnp.atleast_1d(z).shape)<3:
                # here `z` is either:
                # - a single value (), 
                # - a 1d array representing the samples of 1 event (nsamples, )
                # - a 2d array representing the samples of many events (nevents, nsamples)
                # output has shape (nparams, ), (nparams, nsamples) or (nparams, nevents, nsamples)
                return jax.vmap(dVc_at_z, in_axes=(None,0,0,0,0,0,0,None))(z, *self.lambda_c.values(), distances)
            else:
                # here `z` is a 3d array representing the samples of many events
                # obtained for different values of the cosmological parameters (nparams, nevents, nsamples)
                # need to vectorize the first axis of `z` *with* cosmo params arrays
                # output has same shape of `z`
                return jax.vmap(dVc_at_z, in_axes=(0,0,0,0,0,0,0,None))(z, *self.lambda_c.values(), distances)
        return dVc_at_z(z,*self.lambda_c.values(), distances)
    
    def V_at_z(self, z, distances=None):
        if self.vectorize_computation:
            if len(jnp.atleast_1d(z).shape)<3:
                return jax.vmap(Vc_at_z, in_axes=(None,0,0,0,0,0,0,None))(z, *self.lambda_c.values(), distances)
            else:
                return jax.vmap(Vc_at_z, in_axes=(0,0,0,0,0,0,0,None))(z, *self.lambda_c.values(), distances)
        return Vc_at_z(z,*self.lambda_c.values(), distances)
    
    def ddGW_at_z(self, z, distances=None):
        if self.vectorize_computation:
            if len(jnp.atleast_1d(z).shape)<3:
                return jax.vmap(ddL_at_z, in_axes=(None,0,0,0,0,0,0,None))(z, *self.lambda_c.values(), distances)
            else:
                return jax.vmap(ddL_at_z, in_axes=(0,0,0,0,0,0,0,None))(z, *self.lambda_c.values(), distances)
        return ddL_at_z(z,*self.lambda_c.values(), distances)
    
    def compute_z_grids(self, events_dL, cosmo_prior, z_conf_range, z_int_res):
        
        if isinstance(z_conf_range, list):
            dL_min, dL_max = jnp.percentile(events_dL, z_conf_range, axis=1)
        elif isinstance(z_conf_range, int):
            mu     = jnp.mean(events_dL, axis=1)
            sig    = jnp.std(events_dL, axis=1)
            dL_min = mu - z_conf_range * sig
            dL_max = mu + z_conf_range * sig
        else:
            med    = jnp.median(events_dL)
            mad    = jnp.median(jnp.abs(events_dL - med))
            dL_min = med - z_conf_range * mad
            dL_max = med + z_conf_range * mad

        dL_min = jnp.where(dL_min < 1.e-6, 1.e-6, dL_min)
        
        lc_low  = {"H0" :cosmo_prior['H0'][0], 
                   "Om0":cosmo_prior['Om0'][0], 
                   "Ok0":cosmo_prior['Ok0'][0], 
                   "Or0":cosmo_prior['Or0'][0], 
                   "w0" :cosmo_prior['w0'][0], 
                   "wa" :cosmo_prior['wa'][0]}
        
        lc_high = {"H0" :cosmo_prior['H0'][1], 
                   "Om0":cosmo_prior['Om0'][1], 
                   "Ok0":cosmo_prior['Ok0'][1], 
                   "Or0":cosmo_prior['Or0'][1], 
                   "w0" :cosmo_prior['w0'][1], 
                   "wa" :cosmo_prior['wa'][1]}
        
        self.update_params(lc_low)
        z_min  = self.z_from_dGW(dL_min, max_z = 20., res = 1000)
        
        self.update_params(lc_high)
        z_max  = self.z_from_dGW(dL_max, max_z = 20., res = 1000)

        return jnp.linspace(z_min, z_max, z_int_res, axis=1)

class mg_flrw(flrw):
    def __init__(self, fiducial_lambda_c=None):
        self.name = 'mg_flrw'
        self.lambda_c = { # default values
            'H0': 70.,
            'Om0': 0.3,
            'Ok0': 0.0,
            'Or0': 0.0,
            'w0': -1.0,
            'wa': 0.0,
            'Xi0': 1.0,
            'n':  2.5,
        }
        self.cosmo_keys = self.lambda_c.keys()
        self.vectorize_computation = False
        self.update_fiducial(fiducial_lambda_c)
        logger.info(f'Created `{self.name}` model with fiducial cosmological parameters: {self.fiducial_lambda_c}')
        
    def z_from_dGW(self, dGWs, max_z = 10., res = 250):
        if self.vectorize_computation:
            # dLs will aways have shape (), (nsamples,) or (nevents,nsamples)
            # need only vectorization on cosmo params arrays
            # output has shape (nparams,), (nparams,nsamples) or (nparams,nevents,nsamples)
            return jax.vmap(dGW2z, in_axes=(None,0,0,0,0,0,0,0,0,None,None))(dGWs, *self.lambda_c.values(), max_z, res)
        return dGW2z(dGWs, *self.lambda_c.values(), max_z , res)

    def dVdz_at_z(self, z, distances=None):
        lc = self.lambda_c.copy()
        lc.pop('Xi0')
        lc.pop('n')
        if self.vectorize_computation:
            if len(jnp.atleast_1d(z).shape)<3:
                # here `z` is either:
                # - a single value (), 
                # - a 1d array representing the samples of 1 event (nsamples, )
                # - a 2d array representing the samples of many events (nevents, nsamples)
                # output has shape (nparams, ), (nparams, nsamples) or (nparams, nevents, nsamples)
                return jax.vmap(dVc_at_z, in_axes=(None,0,0,0,0,0,0,None))(z, *lc.values(), distances)
            else:
                # here `z` is a 3d array representing the samples of many events
                # obtained for different values of the cosmological parameters (nparams, nevents, nsamples)
                # need to vectorize the first axis of `z` *with* cosmo params arrays
                # output has same shape of `z`
                return jax.vmap(dVc_at_z, in_axes=(0,0,0,0,0,0,0,None))(z, *lc.values(), distances)
        return dVc_at_z(z,*lc.values(), distances)

    def V_at_z(self, z, distances=None):
        lc = self.lambda_c.copy()
        lc.pop('Xi0')
        lc.pop('n')
        if self.vectorize_computation:
            if len(jnp.atleast_1d(z).shape)<3:
                return jax.vmap(Vc_at_z, in_axes=(None,0,0,0,0,0,0,None))(z, *lc.values(), distances)
            else:
                return jax.vmap(Vc_at_z, in_axes=(0,0,0,0,0,0,0,None))(z, *lc.values(), distances)
        return Vc_at_z(z,*lc.values(), distances)

    def ddGW_at_z(self, z, distances=None):
        if self.vectorize_computation:
            if len(jnp.atleast_1d(z).shape)<3:
                return jax.vmap(ddGW_at_z, in_axes=(None,0,0,0,0,0,0,0,0,None))(z, *self.lambda_c.values(), distances)
            else:
                return jax.vmap(ddGW_at_z, in_axes=(0,0,0,0,0,0,0,0,0,None))(z, *self.lambda_c.values(), distances)
        return ddGW_at_z(z,*self.lambda_c.values(), distances)  
    
    def compute_z_grids(self, events_dL, cosmo_prior, z_conf_range, z_int_res):
        
        if isinstance(z_conf_range, list):
            dL_min, dL_max = jnp.percentile(events_dL, z_conf_range, axis=1)
        elif isinstance(z_conf_range, int):
            mu     = jnp.mean(events_dL, axis=1)
            sig    = jnp.std(events_dL, axis=1)
            dL_min = mu - z_conf_range * sig
            dL_max = mu + z_conf_range * sig
        else:
            med    = jnp.median(events_dL)
            mad    = jnp.median(jnp.abs(events_dL - med))
            dL_min = med - z_conf_range * mad
            dL_max = med + z_conf_range * mad

        dL_min = jnp.where(dL_min < 1.e-6, 1.e-6, dL_min)
        
        lc_low  = {"H0" :cosmo_prior['H0'][0], 
                   "Om0":cosmo_prior['Om0'][0], 
                   "Ok0":cosmo_prior['Ok0'][0], 
                   "Or0":cosmo_prior['Or0'][0], 
                   "w0" :cosmo_prior['w0'][0], 
                   "wa" :cosmo_prior['wa'][0],
                   "Xi0":cosmo_prior['Xi0'][1],
                   "n"  :cosmo_prior['n'][1]}
        
        lc_high = {"H0" :cosmo_prior['H0'][1], 
                   "Om0":cosmo_prior['Om0'][1], 
                   "Ok0":cosmo_prior['Ok0'][1], 
                   "Or0":cosmo_prior['Or0'][1], 
                   "w0" :cosmo_prior['w0'][1], 
                   "wa" :cosmo_prior['wa'][1],
                   "Xi0":cosmo_prior['Xi0'][0],
                   "n"  :cosmo_prior['n'][1] # true if cosmo_prior['Xi0'][0] < 1, otherwise should be cosmo_prior['n'][0]
                  }
        
        self.update_params(lc_low)
        z_min  = self.z_from_dGW(dL_min, max_z = 50., res = 2000)
        
        self.update_params(lc_high)
        z_max  = self.z_from_dGW(dL_max, max_z = 50., res = 2000)

        return jnp.linspace(z_min, z_max, z_int_res, axis=1)
```

```python
cosmo = flrw()

H0 = jnp.linspace(20,200,100)
cosmo.update_params({'H0':H0})

dLscatterH0 = cosmo.z_from_dGW(dLs, max_z=15.)

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as mcolors

ev = 0

# Set up the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Define colormap
cmap = cm.viridis
norm = mcolors.Normalize(vmin=H0.min(), vmax=H0.max())

# Plot each curve with a corresponding color
for p in range(dLscatterH0.shape[0]):
    ax.plot(dLs[ev], dLscatterH0[p, ev, :], color=cmap(norm(H0[p])))

# Create a scalar mappable for the colorbar
sm = cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])

# Add colorbar and set its label
cbar = plt.colorbar(sm, ax=ax)
cbar.set_label('H0 Values')

# Set labels and show the plot
plt.xlabel('dL [Gpc]')
plt.ylabel('z')
plt.title('z(dL) for varying H0')
plt.show()
```
