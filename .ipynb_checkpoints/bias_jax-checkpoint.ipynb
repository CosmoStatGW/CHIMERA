```python
import sys
sys.path.append("/home/mt/softwares/CHIMERA_private/")

from CHIMERA.Likelihood import HyperLike 
from CHIMERA.mass import plp as chim_plp
from CHIMERA.cosmo import flrw as chim_flrw 
from CHIMERA.rate import madau_dickinson as chim_rate

dir_data       = "/home/mt/softwares/CHIMERA_private/data/"

hyperlkl_rec = {# population and cosmo models
               "model_cosmo" :     chim_flrw,
               "model_mass":       chim_plp,
               "model_rate" :      chim_rate,
               # gw event settings
               "data_GW_dir" :     dir_data+"O5_v3/O5_samples_from_fisher_allpars_snrth-25_ieth-0.05_DelOmTh-inf.h5", # gw catalogue directory
               "Nevents" :         None, # if None use all the events in the GW catalogue
               "scale_events":     1, 
               "Nsamples" :        5000, # samples to use for each GW event
               "data_GW_smooth":   0.3,  # KDE bandwidth
               "pixelize":         False, 
               "npix_event":       1,   # mean number of pixels for each event
               "sky_conf":         0.90, # sky area to cover with the pixelization
               "nside_list":       [512,256,128,64,32,16,8], # healpix paramter
               "z_int_sigma":      5,    # standard deviation used to compute z_grids
               "H0_prior":         [20.,200.], # H0 prior used to compute z_grids
               "Xi0_prior":        None, #[0.1,10.0], # Xi0 prior used to compute z_grids. Needed when using mg_flrw
               "n_prior":          None, #[1,5], # n prior used to compute z_grids. Needed when using mg_flrw
               "z_int_res_like":   300,  # redshift grid size for the LikelihoodNumerator 
               "kde_kind":         "approximate", # kde scheme to use
               "kde_cut_sigma":    2,    # sigma outside the dataset support at which stopping computing the kde on the redshift grid 
               "kde_binned" :      False, # option to bin data before computing the kde      
               "num_bins" :        100,   # number of bins in which binning the data before compute the kde
               "shuffle_gw":       False, 
               # catalogue settings 
               "z_det_range":      None, # redshift range of the catalogue
               "data_GAL_zerr":    0.001, # error on the redshift of each galaxy
               "data_GAL_dir":     None,#dir_parent, # galaxy catalogue path
               "p_bkg":            'dV_dz', # kind of background galaxy catalogue to use
               "data_GAL_int_dir": None,#dir_parent_int, # galaxy catalogue interpolant path               
               "band" :            None, # to use for LVK LikelihoodNumerator
               "Lcut" :            None, # to use for LVK LikelihoodNumerator
               "avg_compl_dir":    None, # to use for LVK LikelihoodNumerator
               "completeness":     None, # to use for LVK LikelihoodNumerator
               # bias settings
               "file_inj_dir" :    dir_data+"O5_v3/injections_20M_sources_PLP_v9s2_H1-L1-Virgo-KAGRA-LIGOI_IMRPhenomHM_snr_th-20_dutyfac-1_fmin-10_noiseless.h5", #injection file path
               "z_int_res_bias":   1000, # redshift grid size for the Bias           
               "neff_min":         5, # neff min used both the bias and the likelihood numerator!
               "N_inj"    :        20*10**6, # number of injections to use
               "snr_th" :          20, # minimum SNR threshold
                }

hyper_model = HyperLike(like_numerator_model = 'mock', recipe=hyperlkl_rec)

lc = chim_flrw._fiducial.copy()
lm = chim_plp._fiducial.copy()
lr = chim_rate._fiducial.copy()

dNchim, xichim = hyper_model.bias.compute(lc,lm,lr)
jnp.sum(dNchim)/(20*1e6), xichim
```

```python
from functools import partial
import jax
jax.config.update('jax_platform_name', 'cpu')
jax.config.update("jax_enable_x64", True)
import jax.numpy as jnp

import logging
logger = logging.getLogger(__name__)

class bias():
    def __init__(self,
                 inj_data,
                 inj_prior,
                 cosmo_model, 
                 mass_model,
                 rate_model,
                 N_inj          = None,
                 p_bkg          = None,
                 z_det_range    = None,
                 z_int_res_bias = 1000,
                 neff_inj_min   = 5,
                 Tobs           = 1,
                 ):
    
        self.inj_data    = inj_data
        self.inj_prior   = inj_prior
        self.N_inj       = N_inj 

        self.cosmo_model  = cosmo_model
        self.mass_model   = mass_model
        self.rate_model   = rate_model

        self.p_bkg        = p_bkg if (p_bkg is not None) else self.cosmo_model.dVdz_at_z
        self.use_dVdz     = False if (p_bkg is not None) else True
        
        self.normalized   = False if z_det_range is None else True
        self.Tobs         = Tobs if ~self.normalized else 1.
        self.z_int_res    = z_int_res_bias
        self.z_det_range  = z_det_range

        self.neff_inj_min = neff_inj_min
        self.check_Neff   = True if self.neff_inj_min is not None else False  
    
    def update_hyper_params(self, hyper_params):
        lc = {k:v for k,v in hyper_params.items() if k in self.cosmo_model.cosmo_keys}
        lm = {k:v for k,v in hyper_params.items() if k in self.mass_model.mass_keys}
        lr = {k:v for k,v in hyper_params.items() if k in self.rate_model.rate_keys}
        
        self.cosmo_model.update_params(lc)
        self.mass_model.update_params(lm)
        self.rate_model.update_params(lr)
    
    def compute(self,  hyper_params=None, ret_normed=False):
        
        # need to implement normalization throug Nexp
        
        if hyper_params:
            self.update_hyper_params(hyper_params)
            
        z_inj            = self.cosmo_model.z_from_dGW(self.inj_data['dL'], max_z = 2.)
        m1s_inj, m2s_inj = self.inj_data["m1det"]/(1.+z_inj),  self.inj_data["m2det"]/(1.+z_inj)
        
        dNdtheta = self.Tobs  * self.mass_model.pdf_m1m2(m1s_inj, m2s_inj) *\
                   self.rate_model.compute_rate(z_inj)/(1.+z_inj) *\
                   self.cosmo_model.dVdz_at_z(z_inj, distances=self.inj_data['dL'])/\
                   (self.cosmo_model.ddGW_at_z(z_inj, distances=self.inj_data['dL'])*(1.+z_inj)**2) /\
                   self.inj_prior  
                    
        if ret_normed:
            norm   = self.N_exp()
            dNdtheta /= norm

        xi   = jnp.sum(dNdtheta) / self.N_inj
        
        # manually check neff
        s2 = jnp.sum(dNdtheta**2) / self.N_inj**2  - xi**2 / self.N_inj
        neff = xi**2 / s2
    
        # need to vectorize here over p_param
        if ((neff <= 0) or (neff < self.neff_inj_min)) and self.check_Neff:
            return 0
        
        return  dNdtheta, xi

    def N_exp(self):
        pass
```

```python
import h5py

dir_data       = "/home/mt/softwares/CHIMERA_private/data/"
dir_inj = dir_data+"O5_v3/injections_20M_sources_PLP_v9s2_H1-L1-Virgo-KAGRA-LIGOI_IMRPhenomHM_snr_th-20_dutyfac-1_fmin-10_noiseless.h5"

def load_injection(dir_inj,  snr_cut=None):

        with h5py.File(dir_inj, 'r') as f:
            m1det = f['m1src'][:]*(1+f['z'][:])
            m2det = f['m2src'][:]*(1+f['z'][:])
            dL    = f['dL'][:]
            snr   = f['SNR_net'][:]
            inj_logprior = f['log_p_draw_nospin'][:]
        
        keep = jnp.full(len(inj_logprior), True)

        if snr_cut is not None:
            keep = keep & (snr > snr_cut)

        assert (m1det > 0).all()
        assert (m2det > 0).all()
        assert (dL > 0).all()
        assert (m2det<=m1det).all()

        inj_data = {'m1det':m1det[keep], 'm2det': m2det[keep], 'dL': dL[keep]} 
        
        return inj_data, jnp.exp(inj_logprior[keep])


inj_data, inj_prior = load_injection(dir_inj, 20)

from CHIMERANew.cosmo import flrw
from CHIMERANew.mass import plp
from CHIMERANew.rate import madau_dickinson
from CHIMERANew.catalog import empty_catalog

cosmo_model = flrw()
mass_model  = plp()
rate_model  = madau_dickinson()

cosmo_model.update_params(lc)
mass_model.update_params(lm)
rate_model.update_params(lr)

bias_model = bias(inj_data,
                 inj_prior,
                 cosmo_model, 
                 mass_model,
                 rate_model,
                 N_inj = 20*1e6)

dNjax, xijax = bias_model.compute()
jnp.sum(dNjax)/(20*1e6), xijax
```
