## n-dimensional KDE

```python
from functools import partial
import jax
jax.config.update('jax_platform_name', 'cpu')
import jax.numpy as jnp

@partial(jax.jit, static_argnames=['bw_method', 'in_log'])
def gaussian_kde_nd(dataset, evaluation_grid, weights=None, bw_method=None, in_log=False):
    
    dataset = jnp.atleast_2d(dataset)    
    d_dataset, n_dataset = dataset.shape
    
    points = jnp.atleast_2d(points)
    d_points, n_points = points.shape
        
    if d_points != d_dataset:
        if d_points == 1 and n_points == d_dataset:
            points = points.T
            n_points = points.shape[1]
        else:
            msg = "points have dimension " + str(d_points) + ", dataset has dimension " + str(d_dataset)
            raise ValueError(msg)
        
    if weights is not None:    
        if weights.ndim != 1:
            raise ValueError("`weights` input should be one-dimensional.")
        if len(weights) != n_dataset:
            raise ValueError("`weights` input should be of length n_dataset")
        _weights = weights /jnp.sum(weights)  
    else:
        _weights = jnp.full(n_dataset, 1.0 / n_dataset, dtype=dataset.dtype)
    
    neff = 1.0 / jnp.sum(jnp.power(_weights,2))

    if bw_method == "scott" or bw_method is None:
        factor = jnp.power(neff, -1. / (d_dataset + 4))
    elif bw_method == "silverman":
        factor = jnp.power(neff * (d_dataset + 2) / 4.0, -1. / (d_dataset + 4))
    elif jnp.isscalar(bw_method) and not isinstance(bw_method, str):
        factor = bw_method
    else:
        raise ValueError("`bw_method` should be 'scott', 'silverman', a scalar")
    
    _mean            = jnp.sum(_weights * dataset, axis=1)
    _residual        = (dataset - _mean[:, None])
    _data_covariance = jnp.atleast_2d(jnp.dot(_residual * _weights, _residual.T))
    _data_covariance /= (1 - jnp.sum(_weights ** 2))
    _data_inv_cov    = jnp.linalg.inv(_data_covariance)

    covariance  = _data_covariance * factor**2
    inv_cov     = _data_inv_cov / factor**2
    
    whitening        = jnp.linalg.cholesky(inv_cov)
    points_whitened  = jnp.dot(points.T, whitening)
    dataset_whitened = jnp.dot(dataset.T, whitening)
    log_norm = jnp.sum(jnp.log(jnp.diag(whitening))) - 0.5 * d_dataset * jnp.log(2 * jnp.pi)

    to_ret = gaussian_kernel_nd(in_log, points_whitened, dataset_whitened, _weights, log_norm)
    return to_ret

@partial(jax.jit, static_argnums=0)
def gaussian_kernel_nd(in_log, points, dataset, weights, log_norm):
    
    def _kernel(x_test, x_train, y_train):
        arg = log_norm - 0.5 * jnp.sum(jnp.square(x_train - x_test))
        return jnp.log(y_train) + arg if in_log else y_train * jnp.exp(arg)
    
    def _reduced_kernel(x):
        kernel_values = vmap(_kernel, in_axes=(None, 0, 0))(x, dataset, weights)
        return special.logsumexp(kernel_values) if in_log else jnp.sum(kernel_values)
    
    mapped_kernel = vmap(_reduced_kernel)
    return mapped_kernel(points)
```

## Univerate KDE

```python
from functools import partial
import jax
jax.config.update('jax_platform_name', 'cpu')
jax.config.update("jax_enable_x64", True)
import jax.numpy as jnp

from CHIMERANew.utils import get_neff

@jax.jit
def epan_kernel(u):
    return jnp.where(jnp.abs(u) <= 1, 3/4 * (1 - u**2), 0)

@jax.jit
def gaussian_kernel(u):
    return jnp.exp(-0.5 * u**2) / jnp.sqrt(2 * jnp.pi)


@partial(jax.jit, static_argnames=['num_bins'])
def simple_binning_1d(dataset, weights, num_bins=200):
    # Determine the bin edges
    min_val = jnp.min(dataset)
    max_val = jnp.max(dataset)
    bin_edges = jnp.linspace(min_val, max_val, num_bins + 1)
    
    # Calculate the bin centers
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    
    # Compute bin indices for each data point
    bin_indices = jnp.clip(jnp.floor((dataset - min_val) / (max_val - min_val) * num_bins), 0, num_bins - 1).astype(int)
    
    # Initialize bin counts array
    bin_counts = jnp.zeros(num_bins)
    
    # Use scatter-add to accumulate weights into corresponding bins
    # No need to normalize apparently
    bin_counts = bin_counts.at[bin_indices].add(weights)
    
    return bin_centers, bin_counts

@partial(jax.jit, static_argnames=['kernel', 'bw_method', 'cut', 'binning', 'num_bins', 'weights_neff', 'data_neff'])
def univariate_kde_routine(dataset, 
                           evaluation_grid, 
                           weights=None, 
                           kernel='epan', 
                           bw_method=None, 
                           cut=2, 
                           binning=True, 
                           num_bins=200,
                           weights_neff=-1.,
                           data_neff=-1.):
    # Handle weights
    if weights is None:
        weights = jnp.ones_like(dataset) / dataset.size
    else:
        weights = weights / jnp.sum(weights)

    norms = jnp.mean(weights, axis=-1)
    n_eff = get_neff(weights, norms)

    # Define the valid density computation
    def valid_density_fn(dataset, weights, kernel, bw_method, cut, num_bins, evaluation_grid, binning):
        # Binning
        if binning:
            dataset, weights = simple_binning_1d(dataset, weights, num_bins)

      # Bw selection
        neff = 1.0 / jnp.sum(jnp.power(weights, 2))
        if bw_method == "scott" or bw_method is None:
            bandwidth = jnp.power(neff, -1. / (1 + 4))
        elif bw_method == "silverman":
            bandwidth = jnp.power(neff * (1 + 2) / 4.0, -1. / (1 + 4))
        elif jnp.isscalar(bw_method) and not isinstance(bw_method, str):
            bandwidth = bw_method
        else:
            raise ValueError("bw_method should be 'scott', 'silverman', or a scalar")

        # Set up evaluation grid
        data_min, data_max = jnp.min(dataset), jnp.max(dataset)
        sigma = jnp.std(dataset)
        lb    = data_min - cut*sigma 
        lbb   = jnp.where(lb > 0., lb, 1.e-6)
        grid  = jnp.linspace(lbb, data_max + cut*sigma, evaluation_grid.size // 2)

        # Choose kernel and compute kernel values
        kernel_fn   = epan_kernel if kernel == 'epan' else gaussian_kernel
        kernel_vals = kernel_fn((grid[:, None] - dataset) / bandwidth)

        # Calculate density
        density = jnp.sum(weights * kernel_vals, axis=1) / bandwidth

        # Interpolate density at evaluation_gird
        return jnp.interp(evaluation_grid, grid, density, left = 0., right = 0.)

    # Dynamically select between valid and invalid computation paths
    return jax.lax.cond(
        (jnp.sum(weights, axis=-1) >= weights_neff) & (n_eff >= data_neff),
        lambda _: valid_density_fn(dataset, weights, kernel, bw_method, cut, num_bins, evaluation_grid, binning),
        lambda _: jnp.zeros_like(evaluation_grid), # return zero if check_neff fails
        operand=None
    )


@partial(jax.jit, static_argnames=['kernel', 'bw_method', 'cut', 'binning', 'num_bins', 'weights_neff', 'data_neff'])
def univariate_kde(dataset, 
                   evaluation_grid, 
                   weights=None, 
                   kernel='epan', 
                   bw_method=None, 
                   cut=2, 
                   binning=True, 
                   num_bins=200,
                   weights_neff=-1.,
                   data_neff=-1.):
    n_axis = len(dataset.shape)
    if n_axis == 1:
        return univariate_kde_routine(dataset, evaluation_grid, weights, kernel, bw_method, cut, binning, num_bins, weights_neff, data_neff)
    elif n_axis == 2:
        batched_kde = jax.vmap(univariate_kde_routine, in_axes=(0, 0, 0, None, None, None, None, None, None, None))
        return batched_kde(dataset, evaluation_grid, weights, kernel, bw_method, cut, binning, num_bins, weights_neff, data_neff)
    elif n_axis == 3:
        batched_kde = jax.vmap(jax.vmap(univariate_kde_routine, in_axes=(0, 0, 0, None, None, None, None, None, None, None)),
                               in_axes=(0, None, 0, None, None, None, None, None, None, None))
        return batched_kde(dataset, evaluation_grid, weights, kernel, bw_method, cut, binning, num_bins, weights_neff, data_neff)
    else:
        raise ValueError("Inconsistent shapes of `dataset`, `weights` and `evaluation_grid`.")
```

```python
# Test 1d KDE not vectorized
seed = 11
key  = jax.random.PRNGKey(seed)
data = jax.random.normal(key, shape=(5000,)) + 4
w    = (data-4)**2

grid = jnp.linspace(jnp.min(data)-2, jnp.max(data)+2, 300)

import matplotlib.pyplot as plt

pdf_epan  = univariate_kde(data, grid, weights=w, binning=False)
pdf_gauss = univariate_kde(data, grid, weights=w, kernel='gaussian', binning=False)

pdf_epan_binned  = univariate_kde(data, grid, weights=w, binning=True)
pdf_gauss_binned = univariate_kde(data, grid, weights=w, kernel='gaussian', binning=True, num_bins=1000)

plt.plot(grid, pdf_epan, label='Epan kernel')
plt.plot(grid, pdf_gauss, label='Gaussian kernel')
plt.plot(grid, pdf_epan_binned, label='Epan kernel binned')
plt.plot(grid, pdf_gauss_binned, label='Gaussian kernel binned')
plt.hist(data, weights=w, density=True, bins=40, label='Histogram', alpha=0.5)
plt.legend()
plt.show()
```
