```python
from functools import partial
import jax
jax.config.update('jax_platform_name', 'cpu')
jax.config.update("jax_enable_x64", True)
import jax.numpy as jnp

from CHIMERANew.kde import univariate_kde
from CHIMERANew.utils import get_neff

import logging
logger = logging.getLogger(__name__)

class like_numerator():
    
    def __init__(self, events_pe, 
                       pe_prior, 
                       cosmo_model, 
                       mass_model,
                       rate_model,
                       galcat_model,  
                       # check Neff settings
                       positive_weights_thresh = 5.,  
                       data_Neff    = 5.,
                       # kde settings,
                       kernel    ='epan', 
                       bw_method = None, 
                       cut_grid  = 2, 
                       binning   = True, 
                       num_bins  = 200,                     
                ):
        
        self.events_pe   = events_pe
        self.pe_prior    = pe_prior
        
        self.cosmo_model = cosmo_model
        self.mass_model   = mass_model        
        self.rate_model   = rate_model
        
        self.galcat_model = galcat_model
        self.zgrids = self.galcat_model.zgrids
        self.zmax   = self.galcat_model.zmax
        
        # Neff settings
        self.positive_weights_thresh = positive_weights_thresh
        self.data_Neff    = data_Neff
        
        # KDE settings
        self.kernel    = kernel
        self.bw_method = bw_method
        self.cut_grid  = cut_grid  
        self.binning   = binning   
        self.num_bins  = num_bins  
    
    def update_hyper_params(self, hyper_params):
        lc = {k:v for k,v in hyper_params.items() if k in self.cosmo_model.cosmo_keys}
        lm = {k:v for k,v in hyper_params.items() if k in self.mass_model.mass_keys}
        lr = {k:v for k,v in hyper_params.items() if k in self.rate_model.rate_keys}
        
        self.cosmo_model.update_params(lc)
        self.mass_model.update_params(lm)
        self.rate_model.update_params(lr)

    def compute_pgw(self, hyper_params=None, ret_normed=True):
        
        if hyper_params:
            self.update_hyper_params(hyper_params)

        zs       = self.cosmo_model.z_from_dGW(self.events_pe['dL'], max_z=self.zmax)
        m1s, m2s = self.events_pe['m1det']/(1.+zs), self.events_pe['m2det']/(1.+zs)
        
        weights  = self.mass_model.pdf_m1m2(m1s, m2s)/self.pe_prior
        
        p_gw = univariate_kde(zs, 
                              self.zgrids, 
                              weights,
                              kernel=self.kernel, 
                              bw_method=self.bw_method, 
                              cut=self.cut_grid, 
                              binning=self.binning, 
                              num_bins=self.num_bins,
                              positive_weights_thresh=self.positive_weights_thresh,
                              data_neff=self.data_Neff)
        
        if ret_normed:
            norms = jnp.mean(weights, axis = 1)
            p_gw *= norms[:,jnp.newaxis]
        
        return p_gw
                
    def compute_pgw_pixelized(self, hyper_params=None, ret_normed=True):
        pass # to be implemented  
        
    def compute(self, hyper_params=None):
        
        if hyper_params:
            self.update_hyper_params(hyper_params)
        
        pgw   = self.compute_pgw(ret_normed=True)
        pgal  = self.galcat_model.compute_pgal()
        prate = self.rate_model.compute_rate(self.zgrids) 
        
        jacobian = (1.+self.zgrids)**3 * self.cosmo_model.ddGW_at_z(self.zgrids)
        
        pz       = pgw*pgal*prate / jacobian
        
        like_evs = jnp.trapz(pz, self.zgrids)
        
        return jnp.nansum(like_evs, axis = 0)
```
