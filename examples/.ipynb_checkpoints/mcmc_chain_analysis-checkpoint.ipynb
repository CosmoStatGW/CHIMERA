## 1. If you restarted an MCMC chain, and so you have mutiple h5 files, you first have to concatenate them into a unique h5 file that can be read by emcee

```python
import h5py
import numpy as np
import tempfile

def remove_trailing_zeros(dataset):
    """Remove trailing zeros from a 2D numpy array along the first axis."""
    return dataset[:np.max(np.nonzero(dataset[:, 0])) + 1]

def conc_two_files(file1, file2, conc_file, burnin1, burnin2):
    with h5py.File(file1, 'r') as f1, h5py.File(file2, 'r') as f2:
        # Create the group
        group_name = 'mcmc'
        conc_group = conc_file.create_group(group_name)

        # Copy the attributes from the first file
        for attr_name, attr_value in f1[group_name].attrs.items():
            conc_group.attrs[attr_name] = attr_value

        # Concatenate the datasets from both files
        for dataset_name in ['chain', 'log_prob']:
            dataset1 = remove_trailing_zeros(f1[group_name][dataset_name][burnin1:])
            dataset2 = remove_trailing_zeros(f2[group_name][dataset_name][burnin2:])
            conc_dataset = np.concatenate((dataset1, dataset2), axis=0)
            conc_group.create_dataset(dataset_name, data=conc_dataset)

        # Sum the 'accepted' dataset along axis=0
        accepted1 = f1[group_name]['accepted'][:]
        accepted2 = f2[group_name]['accepted'][:]
        concatenated_accepted = np.sum((accepted1, accepted2), axis=0)
        conc_group.create_dataset('accepted', data=concatenated_accepted)

        # Update the 'iteration' attribute
        iteration = f1[group_name].attrs.get('iteration', 0) + f2[group_name].attrs.get('iteration', 0)
        conc_group.attrs['iteration'] = iteration

def conc_chains(files_list, output_file, burnin=None):
    if len(files_list) < 2:
        print("Need at least two files to concatenate. Returning.")
        return None

    if burnin is None:
        burnin = [0] * len(files_list)

    with h5py.File(output_file, 'w') as conc_file:
        # Base case: concatenate the last two files in the list
        if len(files_list) == 2:
            conc_two_files(files_list[0], files_list[1], conc_file, burnin1=burnin[0], burnin2=burnin[1])
            return output_file
        else:
            # Recursive case: concatenate the first file with the concatenated output of the rest
            with tempfile.NamedTemporaryFile() as temp_file:
                conc_two_files(files_list[0], conc_chains(files_list[1:], temp_file.name, burnin=burnin[1:]), conc_file, burnin1=burnin[0], burnin2=0)
            return output_file
        

files_to_conc = ["chain_0.h5", 
                 "chain_1.h5"]

output_file = "chain_conc.h5"  # I suggest the same prefix without the final numbers 

conc_chains(files_to_conc, output_file) 
```

```python
import emcee

output_file1 = "chain_0.h5"

reader1 = emcee.backends.HDFBackend(output_file, read_only=True)


flatchain1 = reader1.get_chain(flat=True) 
chain1     = reader1.get_chain()
log_prob1  = reader1.get_log_prob()

print(chain1.shape, flatchain1.shape)
```

```python
from emcee.autocorr import integrated_time

try:
    tau = reader1.get_autocorr_time(quiet=True)
    burnin = int(2 * np.max(tau))
    thin = int(0.5 * np.min(tau))
except:
    tau = reader1.get_autocorr_time(quiet=True)
    
```
